```python
predicate_list = []
object_list = []

# Iterate through the column
for row in unique_interaction_action_df['InteractionActions']:
    # Split by '|' to get individual pairs
    pairs = row.split('|')
    
    for pair in pairs:
        # Split by '^' to separate predicate and object
        # Using maxsplit=1 is safer if the object contains a '^'
        if '^' in pair:
            pred, obj = pair.split('^', 1)
            predicate_list.append(pred)
            object_list.append(obj)

# Displaying the first few results
print(f"Predicates: {predicate_list[:5]}")
print(f"Objects: {object_list[:5]}")

print("predicate_list count:", len(predicate_list))
print("object_list count:", len(object_list))
```

```python
unique_predicates = sorted(list(set(predicate_list)))
unique_objects = sorted(list(set(object_list)))

print("unique_predicates count:", len(unique_predicates))
print("unique_objects count:", len(unique_objects))
```

<h3>I noticed in the dataset that sometimes </br> 
    "increases" (from column InteractionActions') appears as "increased" (from column 'Interaction') </br> 
    or "decreases" (from column InteractionActions') appears as "decreased" (from column 'Interaction').</h3>

<h3>It is necessary to handle "synonyms"</h3>

```python
import re

def check_actions_in_interaction(row):
    # 1. Normalize and handle missing data
    action_str = str(row['InteractionActions']).lower()
    interaction_text = str(row['Interaction']).lower()
    
    # 2. SMART Concept Groups
    # We map formal categories to specific biological synonyms and active verbs.
    concept_map = {
        # General Affects: Includes binding, inhibition, and general outcomes
        "affects": ["inhibits", "promotes", "induces", "activates", "modulates", "results in", "susceptibility", "affects", "binds", "impacts", "acts on"],
        
        # Binding: Covers various states of physical interaction
        "binding": ["binds", "bound", "binding", "complex", "associates", "interaction", "ligand", "docking", "attachment"],
        
        # Directional Changes: Standardizing increase/decrease synonyms
        "decreases": ["decreased", "decreases", "inhibits", "reduction", "lower", "attenuates", "blocks", "suppresses", "downregulates", "diminishes", "depletes"],
        "increases": ["increased", "increases", "promotes", "induction", "higher", "upregulates", "activates", "stimulates", "augments", "elevates"],
        
        # Metabolic & Chemical Processes
        "metabolic processing": ["metabolism", "metabolized", "metabolic", "processing", "catabolism", "anabolism", "biotransformation", "degradation"],
        "chemical synthesis": ["synthesis", "synthesized", "production", "formation", "generation", "synthesizing"],
        "reaction": ["reaction", "process", "interaction", "transformation", "conversion", "pathway"],
        
        # Outcome Measures: Abundance, Expression, Activity
        "expression": ["expression", "levels", "mrna", "protein levels", "transcript", "expressed"],
        "abundance": ["abundance", "levels", "concentration", "amount", "quantity", "presence"],
        "activity": ["activity", "active", "function", "catalytic", "enzymatic", "action"],
        
        # Clinical & Experimental Terms
        "response to substance": ["susceptibility", "sensitivity", "resistance", "response", "reactivity", "effect of"],
        "cotreatment": ["co-treated", "cotreatment", "combination", "together with", "presence of", "addition of"],
        
        # Specific Biological Modifications
        "cleavage": ["cleavage", "cleaved", "proteolysis", "hydrolysis", "breakdown", "fragmentation"],
        "phosphorylation": ["phosphorylation", "phosphorylated", "phospho", "kinase activity"]
    }

    # 3. Extract unique terms from the InteractionActions (splitting by ^ and |)
    parts = set(re.split(r'[\^|]', action_str))
    
    def verify_part(part, text):
        part = part.strip()
        if not part: return True
        
        # Step A: Direct Substring or Root match (handles 'cleavage' vs 'cleaved')
        if part in text: return True
        
        # Strip common suffixes to find the linguistic root
        root = re.sub(r'(es|s|ed|ing|ion|ism|ity)$', '', part)
        if len(root) > 3 and root in text: return True
        
        # Step B: Concept Map / Synonym Check
        # Check both the full part and its root in the map
        for key in [part, root]:
            if key in concept_map:
                if any(synonym in text for synonym in concept_map[key]):
                    return True
        return False

    # 4. Final logic check: all parts of the Action string must be validated
    return all(verify_part(p, interaction_text) for p in parts if p.strip())
```

```python
# using df_human_proteins as the source
actions_series = df_human_proteins.apply(check_actions_in_interaction, axis=1)

# Build the new dataframe by selecting columns AND adding the new series
result_df_actions = df_human_proteins[['InteractionActions', 'Interaction', 'ChemicalName', 'GeneSymbol']].copy()
result_df_actions['ActionsPresent'] = actions_series

# Reorder columns so 'ActionsPresent' is first
cols = ['ActionsPresent', 'InteractionActions', 'Interaction', 'ChemicalName', 'GeneSymbol']
result_df_actions = result_df_actions[cols]

# Display results
print(result_df_actions.head())
```
