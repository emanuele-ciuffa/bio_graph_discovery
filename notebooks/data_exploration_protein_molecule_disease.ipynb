```python
import pandas as pd
```

<h2>Molecule-Protein Interaction Dataset</h2>

```python
col_names_molecule_protein = ["ChemicalName", "ChemicalID", "CasRN", "GeneSymbol", "GeneID", "GeneForms", "Organism", "OrganismID", "Interaction", "InteractionActions", "PubMedIDs"]
ds_molecule_protein = pd.read_csv("data/CTD_chem_gene_ixns/CTD_chem_gene_ixns.csv", names=col_names_molecule_protein, sep=",", header=None)

print(ds_molecule_protein.head())
```

```python
print("ds_molecule_protein len: ", len(ds_molecule_protein))
```

```python
ds_human = ds_molecule_protein[ds_molecule_protein["Organism"] == "Homo sapiens"]
print("ds_human len: ", len(ds_human))
```

<h3>GENE FORMS</hr>

```python
# Cleans up whitespace and makes everything uppercase before finding unique values
gene_types = ds_human["GeneForms"].str.strip().str.upper().unique()

print(gene_types)
```

<h2>protein exploration in dataset Molecule-Protein Interaction</h3>

```python
# Create the filtered dataframe
ds_protein = ds_human[ds_human["GeneForms"] == "protein"]
```

```python
symbols = ds_protein["GeneSymbol"].unique()
print("PROTEINS IN RECORDS: ", ds_protein["GeneSymbol"].count())
print("SYMBOLS: ", symbols)
print("DISTINCT PROTEIN SYMBOLS: ", len(symbols))
```

```python
print(ds_protein["GeneSymbol"].value_counts())
```

```python
target_forms = ['protein', 'modified form', 'mutant form', 'alternative form']
ds_protein_all = ds_human[ds_human["GeneForms"].isin(target_forms)]

print(ds_protein_all["GeneForms"].value_counts())
```

<h2>molecule exploration in dataset Molecule-Protein Interaction</h2>

```python
df_molecule = ds_human["ChemicalName"].unique()

print("MOLECULES: ", df_molecule)
print("\nDISTINCT MOLECULES: ", len(df_molecule))
```

<h2>Molecule-Disease Dataset</h2>

```python
col_names_molecule_disease = ["ChemicalName","ChemicalID","CasRN","DiseaseName","DiseaseID","DirectEvidence","PubMedIDs"]
molecule_disease_ds = pd.read_csv("data/CTD_curated_chemicals_diseases/CTD_curated_chemicals_diseases.csv", names=col_names_molecule_disease, sep=",", header=None)

print(molecule_disease_ds.head)
```

```python
print("molecule_disease_ds count: ", len(molecule_disease_ds))
```

<h3>molecule exploration in dataset Molecule-Disease</h3>

```python
df_molecule2 = molecule_disease_ds["ChemicalName"].unique()

print("MOLECULES: ", df_molecule2)
print("\nDISTINCT MOLECULES: ", len(df_molecule2))
```

```python
import numpy as np

# Find intersection directly on the arrays
intersection = np.intersect1d(df_molecule, df_molecule2)

print("COMMON MOLECULES: ", intersection)
print("\nCOUNT OF COMMON MOLECULES: ", len(intersection))
```

<h3>disease exploration in dataset Molecule-Disease</h3>

```python
df_diseases = molecule_disease_ds["DiseaseName"].unique()

print("DISEASES: ", df_diseases)
print("\nDISTINCT DISEASES: ", len(df_diseases))
```

```python
gouped_by_molecule_disease = molecule_disease_ds[["ChemicalName","DiseaseName"]].value_counts()
print(gouped_by_molecule_disease)
```

```python
print("GROUPED MOLECULE-DISEASE:", len(gouped_by_molecule_disease))
```
