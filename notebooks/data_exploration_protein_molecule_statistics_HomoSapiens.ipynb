```python
# Organism
organism = "Homo sapiens"

statistics = {}
statistics["00_organism"] = organism
```

<h2>Molecule-Protein Interaction Dataset</h2>

```python
col_names_molecule_protein = ["ChemicalName", "ChemicalID", "CasRN", "GeneSymbol", "GeneID", "GeneForms", "Organism", "OrganismID", "Interaction", "InteractionActions", "PubMedIDs"]
ds_molecule_protein = pd.read_csv("data/CTD_chem_gene_ixns/CTD_chem_gene_ixns.csv", names=col_names_molecule_protein, sep=",", header=None)

print(ds_molecule_protein.head())
```

```python
print("ds_molecule_protein len: ", len(ds_molecule_protein))
statistics["01_total_dataset_records"] = len(ds_molecule_protein)
```

```python
ds_organism = ds_molecule_protein[ds_molecule_protein["Organism"] == organism]
print("ds_organism len: ", len(ds_organism))
statistics["02_records_for_organism"] = len(ds_organism)
organism_percentage_for_dataset = str(round(statistics["02_records_for_organism"] / statistics["01_total_dataset_records"] * 100, 2)) + "%"
print("organism_percentage_for_dataset", organism_percentage_for_dataset)
statistics["03_organism_percentage_for_dataset"] = organism_percentage_for_dataset
```

<h3>GENE FORMS</hr>

```python
# Cleans up whitespace and makes everything uppercase before finding unique values
gene_types = ds_organism["GeneForms"].str.strip().str.upper().unique()

print(gene_types)
```

<h2>protein exploration in dataset Molecule-Protein Interaction</h3>

```python
# Create the filtered dataframe
ds_protein = ds_organism[ds_organism["GeneForms"] == "protein"]
```

```python
symbols = ds_protein["GeneSymbol"].unique()
print("PROTEINS IN RECORDS: ", ds_protein["GeneSymbol"].count())
print("SYMBOLS: ", symbols)
print("DISTINCT PROTEIN SYMBOLS: ", len(symbols))
statistics["04_distinct_proteins"] = len(symbols)
```

```python
print(ds_protein["GeneSymbol"].value_counts())
```

```python
target_forms = ['protein', 'modified form', 'mutant form', 'alternative form']
ds_protein_all = ds_organism[ds_organism["GeneForms"].isin(target_forms)]

print(ds_protein_all["GeneForms"].value_counts())
```

```python
print("molecule_disease_ds count: ", len(molecule_disease_ds))
```

<h3>molecule exploration in dataset Molecule-Disease</h3>

```python
df_molecule2 = molecule_disease_ds["ChemicalName"].unique()

print("MOLECULES: ", df_molecule2)
print("\nDISTINCT MOLECULES: ", len(df_molecule2))
```

```python
import numpy as np

# Find intersection directly on the arrays
intersection = np.intersect1d(df_molecule, df_molecule2)

print("COMMON MOLECULES: ", intersection)
print("\nCOUNT OF COMMON MOLECULES: ", len(intersection))
```

```python
df_moleculeID = ds_organism["ChemicalID"].unique()

print("MOLECULES: ", df_moleculeID)
print("\nDISTINCT MOLECULES: ", len(df_moleculeID))
statistics["06_distinct_molecules"] = len(df_moleculeID)
```

```python
df_moleculeID2 = molecule_disease_ds["ChemicalID"].unique()

print("MOLECULES: ", df_moleculeID2)
print("\nDISTINCT MOLECULES: ", len(df_moleculeID2))
```

```python
# intersection by molecule ID (ChemicalID)
intersectionID = np.intersect1d(df_moleculeID, df_moleculeID2)

# Filter df_diseases where "ChemicalID" is in the intersection
intersaction_mol_ID = np.intersect1d(df_moleculeID, df_moleculeID2)

print("COMMON MOLECULES ID: ", intersaction_mol_ID)
print("\nCOUNT OF COMMON MOLECULES ID: ", len(intersaction_mol_ID))

statistics["07_molecules_instersaction_2datasets"] = len(intersaction_mol_ID)
```
