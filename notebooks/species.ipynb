```python
import pandas as pd
```

```python
col_names_molecule_protein = ["ChemicalName", "ChemicalID", "CasRN", "GeneSymbol", "GeneID", "GeneForms", "Organism", "OrganismID", "Interaction", "InteractionActions", "PubMedIDs"]
ds_molecule_protein = pd.read_csv("data/CTD_chem_gene_ixns/CTD_chem_gene_ixns.csv", names=col_names_molecule_protein, sep=",", header=None)
```

```python
species_df = ds_molecule_protein["Organism"].drop_duplicates()
species_df = species_df.reset_index()
print(species_df)
```

```python
# Aggregate the counts for each organism and convert to a DataFrame
organism_df = ds_molecule_protein['Organism'].value_counts().reset_index()

# Rename the columns for clarity
organism_df.columns = ['Organism', 'RecordCount']

# Display the resulting dataframe
print(organism_df)

organism_df.to_csv("organisms.csv", index=False)
```
